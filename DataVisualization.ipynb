```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "task3.csv"
data = pd.read_csv(file_path)
```

```python
category_scores = data.groupby('Category')['Score'].sum().reset_index()
top_categories = category_scores.sort_values(by='Score', ascending=False).head(5)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_categories['Category'], top_categories['Score'], color='skyblue')
plt.title('Top 5 Most Popular Content Categories', fontsize=14)
plt.xlabel('Content Category', fontsize=12)
plt.ylabel('Total Score', fontsize=12)
plt.xticks(rotation=45)
plt.show()
plt.savefig("Top5.png")
```

```python
data["Datetime"] = pd.to_datetime(data["Datetime"], format="%d/%m/%y %H:%M")

# Extract the month from the 'Datetime' column
data["Month"] = data["Datetime"].dt.month

# Count the number of posts by month
posts_by_month = data["Month"].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(10, 6))
posts_by_month.plot(kind="bar", color="skyblue", edgecolor="black")

# Add titles and labels
plt.title("Number of Posts by Month", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Posts", fontsize=12)
plt.xticks(ticks=range(12), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"], rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python


# Convert 'Datetime' column to datetime format
data["Datetime"] = pd.to_datetime(data["Datetime"], format="%d/%m/%y %H:%M")

# Extract the month from the 'Datetime' column
data["Month"] = data["Datetime"].dt.month

# Count the number of posts by month
posts_by_month = data["Month"].value_counts().sort_index()

# Get indices of the max and min months
max_month = posts_by_month.idxmax()
min_month = posts_by_month.idxmin()

# Assign colors for bars
bar_colors = ['green' if month == max_month else 'red' if month == min_month else 'skyblue' for month in posts_by_month.index]

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(posts_by_month.index, posts_by_month.values, color=bar_colors, edgecolor="black")

# Customize y-axis range
plt.ylim(1750, posts_by_month.max() + 200)

# Add titles and labels
plt.title("Number of Posts by Month", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Posts", fontsize=12)
plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"], rotation=45)

# Add annotations for max and min months
plt.text(max_month, posts_by_month[max_month] + 50, "Highest", color="green", ha="center", fontsize=10)
plt.text(min_month, posts_by_month[min_month] + 50, "Lowest", color="red", ha="center", fontsize=10)

plt.tight_layout()
plt.show()
plt.savefig("month.jpg")
```

```python



# Group by 'Category' and calculate the total score for each category
category_scores = data.groupby("Category")["Score"].sum().reset_index()

# Sort the categories by score in descending order
category_scores = category_scores.sort_values(by="Score", ascending=False)

# Identify the top 5 categories
top_5_categories = category_scores.head(5)["Category"]

# Assign colors to categories (green for top 5, blue for the rest)
bar_colors = ["green" if category in top_5_categories.values else "skyblue" for category in category_scores["Category"]]

# Plot the bar chart
plt.figure(figsize=(12, 8))
bars = plt.bar(category_scores["Category"], category_scores["Score"], color=bar_colors, edgecolor="black")

# Highlight top 5 with annotations
for bar, category in zip(bars, category_scores["Category"]):
    if category in top_5_categories.values:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5000, f'{int(bar.get_height())}', 
                 ha='center', va='bottom', fontsize=10, color="black")

# Customize the chart
plt.title("Score of Each Category (Top 5 Highlighted)", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Total Score", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Save the figure
output_path = "category_scores_with_highlighted_top5.png"
plt.savefig(output_path, dpi=300, bbox_inches="tight")  # Save as a high-quality image

# Show the plot for verification
plt.show()

print(f"Figure saved successfully as {output_path}")
```
