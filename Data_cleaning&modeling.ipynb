#### Load Datasets

```python
# Step 1: Load the datasets
reaction_df = dfr
content_df = dfc
reaction_type_df = df
```

#### Merging the datasets

```python
# Step 2: Merge the datasets
# Merge Reaction with Content
merged_df = pd.merge(reaction_df, content_df[['Content ID', 'Category', 'ContentType']], on='Content ID', how='inner')

# Merge with ReactionTypes
merged_df = pd.merge(merged_df, reaction_type_df[['ReactionType', 'Score']], on='ReactionType', how='inner')

merged_df.drop(columns=["Datetime"],inplace=True)

merged_df
```

#### Grouping by Category

```python
# Step 3: Group and summarize for the top 5 categories
# Group by Category and calculate the total score
category_scores = merged_df.groupby('Category')['Score'].sum().reset_index()

# Sort by Score in descending order
category_scores = category_scores.sort_values(by='Score', ascending=False)

category_scores
```

# Top 5 Categories

```python
# Get the top 5 categories
top_5_categories = category_scores.head(5)
top_5_categories
```

# Saving to csv and excel sheets

```python
# Step 4: Export the datasets
# Export to Excel with two sheets
with pd.ExcelWriter('final_dataset.xlsx') as writer:
    merged_df.to_excel(writer, sheet_name='Cleaned Dataset', index=False)
    top_5_categories.to_excel(writer, sheet_name='Top 5 Categories', index=False)

# Optional: Save as separate CSV files
merged_df.to_csv('cleaned_dataset.csv', index=False)
top_5_categories.to_csv('top_5_categories.csv', index=False)
```
